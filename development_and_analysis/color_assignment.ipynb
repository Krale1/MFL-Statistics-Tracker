```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
image_path = "../output_videos/cropped_img.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

Take the top half of the image, where the shirt is

```python
top_half_image = image[:image.shape[0]//2, :, :]
plt.imshow(top_half_image)
```

Cluster the image in 2, the field and the shirt

```python
#Reshape the image to a 2D array of pixels
image_2d = top_half_image.reshape((-1, 3))

#preform k-means clustering with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42).fit(image_2d)
#get the labels of the clusters
labels = kmeans.labels_

#Reshape the labels to the original image shape
labels_reshaped = labels.reshape(top_half_image.shape[0], top_half_image.shape[1])

clustered_top_half_image = labels_reshaped.reshape(top_half_image.shape[0], top_half_image.shape[1])

plt.imshow(clustered_top_half_image)
```

```python
corners_clusters = [clustered_top_half_image[0,0], clustered_top_half_image[0,-1], clustered_top_half_image[-1,0], clustered_top_half_image[-1,-1]]
non_player_cluster = corners_clusters = max(set(corners_clusters), key=corners_clusters.count)
player_cluster = 1-non_player_cluster
player_cluster
```

```python
kmeans.cluster_centers_[player_cluster] 
```
